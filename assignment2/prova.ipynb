```python
import numpy as np
import matplotlib.pyplot as plt

    
path_ = "C:/Users/user.LAPTOP-G27BJ7JO/Desktop/SISSA/CREPALDI/assignment2/drc/fetch.drc/fetch.acts"
```

```python
def extract_float(line: str) -> float:
    
    for token in line.split():
        try:
            float_token = float(token)
            return float_token
        except ValueError:
            pass
```

```python
activations = {
    "TL":    [],
    "TGPC":  [],
    "TPh":   [],
    "TP":    [],
    "TGPCR": []
}

activations_checks = ["TGPC ", "TP ", "TPh", "TGPCR"]

cycles = 0

with open(path_, "r") as f:
    
        for line in f:
            
            float_line = extract_float(line=line)
            
            if "TL" in line:
                
                cycles += 1
                
                activations["TL"].append(float_line)
                
                for a in activations_checks:
                    
                    if len(activations[a.strip()]) == cycles-2:
                        activations[a.strip()].append(0)
                        
            else:
                for a in activations_checks:
                    if a in line:
                        activations[a.strip()].append(float_line)            
            

print(len(activations["TL"]))  
print(len(activations["TGPC"]))   
print(len(activations["TGPCR"]))   
print(len(activations["TP"]))   
print(len(activations["TPh"]))   
```
